```python
# app.py
import streamlit as st
import pandas as pd
import joblib

# Load trained model
model = joblib.load("financial_inclusion_model.pkl")

st.title("Financial Inclusion Prediction (East Africa)")

st.write("Fill in your details to predict if an individual has a bank account.")

# Input fields
age = st.number_input("Age", min_value=0, max_value=120, value=25)
gender = st.selectbox("Gender", ["Male", "Female"])
marital_status = st.selectbox("Marital Status", ["Single", "Married", "Other"])
education = st.selectbox("Education Level", ["None", "Primary", "Secondary", "Tertiary"])
job_type = st.selectbox("Job Type", ["Unemployed", "Self-employed", "Salaried", "Other"])

# Convert input to DataFrame
input_dict = {
    "age": [age],
    "gender": [gender],
    "marital_status": [marital_status],
    "education": [education],
    "job_type": [job_type]
}

input_df = pd.DataFrame(input_dict)

# One-hot encode categorical variables
categorical_cols = ['gender', 'marital_status', 'education', 'job_type']
input_df = pd.get_dummies(input_df, columns=categorical_cols, drop_first=True)

# Ensure columns match model
for col in model.feature_names_in_:
    if col not in input_df.columns:
        input_df[col] = 0
input_df = input_df[model.feature_names_in_]

# Prediction
if st.button("Predict"):
    prediction = model.predict(input_df)
    st.write(f"Prediction: {'Has Bank Account' if prediction[0]==1 else 'Does Not Have Bank Account'}")
```
